```python
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy.stats import binom
```

```python
# Plotting Binomial Distribution WITHOUT using inbuilt functions

# Parameters for the Binomial distribution
n = 10        # number of trials
p = 0.5       # probability of success

# Generate possible outcomes (0 to n)
x = np.arange(0, n + 1)

# Calculate PMF
# P(X=k) = (nCk) * (p^k) * (1-p)^(n-k)
def factorial(num):
    return 1 if num == 0 else num * factorial(num - 1)

def nCr(n, r):
    return factorial(n) / (factorial(r) * factorial(n - r))

# Compute probabilities
y = [nCr(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in x]

# Plotting the Binomial Distribution (Manual)
plt.bar(x, y, color='purple')
plt.title('Binomial Distribution (Without Inbuilt Function)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability P(X=k)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Plotting Binomial Distribution USING inbuilt functions

# Using scipy.stats.binom
n = 10
p = 0.5
x = np.arange(0, n + 1)

# Calculate PMF using built-in function
pmf = binom.pmf(x, n, p)

# Plotting the Binomial Distribution (Inbuilt)
plt.bar(x, pmf, color='brown')
plt.title('Binomial Distribution (Inbuilt Function)')
plt.xlabel('Number of Successes (k)')
plt.ylabel('Probability P(X=k)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
