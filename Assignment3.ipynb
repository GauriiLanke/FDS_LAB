```python
#Name: Gauri Lanke
#Roll NO: 150
#Division: B


import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Name': ['Gauri', 'Payal', 'Shraddha', 'Sakshi', 'Rohit', 'Shreyas', 'Virat', 'Hardik'],  # Qualitative
    'Patient ID': [101, 102, 103, 104, 105, 106, 107, 108],  # Quantitative
    'Gender': ['Female', 'Female', 'Female', 'Female', 'Male', 'Male', 'Male', 'Male'],  # Qualitative
    'Doctor Assigned': ['Dr. Mehta', 'Dr. Sharma', 'Dr. Gupta', 'Dr. Mehta', 'Dr. Sharma', 'Dr. Gupta', 'Dr. Mehta', 'Dr. Gupta'],  # Qualitative
    'Heart Rate (bpm)': [78, 85, 90, 76, 88, 92, 80, 84],  # 
    'SpO2 (%)': [98, 97, 96, 99, 95, 94, 97, 98],  # 
    'O2 Count (L/min)': [2.5, 3.0, 2.8, 2.4, 3.2, 3.5, 2.9, 3.1],  # 
    'Age': [19, 20, 18, 22, 17, 23, 22, 25]  #
}

df = pd.DataFrame(data)
print(df)
```

```python
# --- QUALITATIVE ANALYSIS ---
print("Qualitative Analysis:")

# Frequency of Gender
print("\nGender Frequency:")
print(df['Gender'].value_counts())

# Frequency of Doctor Assigned
print("\nDoctor Assigned Frequency:")
print(df['Doctor Assigned'].value_counts())

# Bar chart for Gender distribution
df['Gender'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

```

```python
# Bar chart for Doctor Assigned distribution
df['Doctor Assigned'].value_counts().plot(kind='bar', color='lightgreen')
plt.title("Doctor Assigned Distribution")
plt.xlabel("Doctor")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Histogram for Heart Rate
df['Heart Rate (bpm)'].plot(kind='hist', bins=5, color='orange', edgecolor='black')
plt.title("Heart Rate Distribution")
plt.xlabel("Heart Rate (bpm)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# --- QUANTITATIVE ANALYSIS ---
print("\nQuantitative Analysis:")
print(df[['Patient ID', 'Heart Rate (bpm)', 'SpO2 (%)', 'O2 Count (L/min)', 'Age']].describe())


# Histogram for SpO2
df['SpO2 (%)'].plot(kind='hist', bins=5, color='purple', edgecolor='black')
plt.title("SpO2 Distribution")
plt.xlabel("SpO2 (%)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Histogram for O2 Count
df['O2 Count (L/min)'].plot(kind='hist', bins=5, color='green', edgecolor='black')
plt.title("O2 Count Distribution")
plt.xlabel("O2 Count (L/min)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Histogram for Age
df['Age'].plot(kind='hist', bins=5, color='blue', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Heart Rate vs. SpO2
plt.scatter(df['Heart Rate (bpm)'], df['SpO2 (%)'], color='blue', edgecolor='black')
plt.title("Heart Rate vs. SpO2")
plt.xlabel("Heart Rate (bpm)")
plt.ylabel("SpO2 (%)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Age vs. O2 Count
plt.scatter(df['Age'], df['O2 Count (L/min)'], color='darkgreen', edgecolor='black')
plt.title("Age vs. O2 Count")
plt.xlabel("Age")
plt.ylabel("O2 Count (L/min)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
